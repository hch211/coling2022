```python
ids = []
claims = []
texts = []

with open('test_origin.tsv','r') as f:
    for item in f.readlines():
        item = item.strip('\n')
        item_list = item.split('\t')
        assert len(item_list)==3
        ids.append(item_list[0])
        claims.append(item_list[1])
        texts.append(item_list[2])
```

```python
test_df = pd.DataFrame({'id':ids[1:],
                       'claim':claims[1:],
                        'text':texts[1:]
                       })
test_df
```

```python
for i in range(len(test_df.text)):
    if type(test_df.text[i]) != str:
        print(i)
```

# pseudo

```python
stance_pred = pd.read_csv('stance_result.tsv', sep='\t')
premise_pred = pd.read_csv('premise_result.tsv', sep='\t')
pseudo = pd.concat([stance_pred,premise_pred.Premise],axis=1)
train_df = pd.read_csv('train.tsv', sep='\t')
pseudo = pd.concat([train_df,pseudo.rename(columns={'text':'Tweet'})])
df_val = pd.read_csv('validation.tsv', sep='\t')
pseudo_df = pd.concat([pseudo,df_val.drop('Unnamed: 0',axis=1)])
pseudo_df.to_csv('pseudo.tsv', sep='\t')
```

```python
pseudo_df = pd.read_csv('pseudo.tsv', sep='\t')
pseudo_df = pseudo_df.drop('Unnamed: 0',axis=1)
pseudo_df
```
