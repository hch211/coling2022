```python
import pandas as pd
```

# 预处理和拼接

```python
train_class = pd.read_csv('Task1a/train/train_class.tsv', sep='\t', header=None)
train_class.columns = ['id', 'Stance']
classes = pd.read_csv('Task1a/valid/class.tsv', sep='\t', header=None)
classes.columns = ['id', 'Stance']
```

```python
test = pd.read_csv('test_tweets_unannotated.tsv', sep='\t', header=None)
test.columns = ['id', 'Tweet']
```

```python
ids = []
texts = []

with open('test_tweets_unannotated.tsv','r') as f:
    for item in f.readlines():
        item = item.strip('\n')
        item_list = item.split('\t')
        assert len(item_list)==2
        ids.append(item_list[0])
        texts.append(item_list[1])
        
test = pd.DataFrame({'id':ids[0:],
                        'Tweet':texts[0:]
                       })
test
```

```python
test.to_csv('test_tweets_unannotated.tsv', sep='\t')
```

```python
ids = []
texts = []

with open('Task1a/train/train_tweets.tsv','r') as f:
    for item in f.readlines():
        item = item.strip('\n')
        item_list = item.split('\t')
        assert len(item_list)==2
        ids.append(item_list[0])
        texts.append(item_list[1])
        
train_tweet = pd.DataFrame({'id':ids[0:],
                        'Tweet':texts[0:]
                       })
train_tweet
```

```python
ids = []
texts = []

with open('Task1a/valid/tweets.tsv','r') as f:
    for item in f.readlines():
        item = item.strip('\n')
        item_list = item.split('\t')
        assert len(item_list)==2
        ids.append(item_list[0])
        texts.append(item_list[1])
        
tweets = pd.DataFrame({'id':ids[0:],
                        'Tweet':texts[0:]
                       })
tweets
```

```python
train_data = pd.merge(train_tweet, train_class, how='left')
valid_data = pd.merge(tweets, classes, how='left')
valid_data
```

```python
train_data.to_csv('Task1a/train/train_data.tsv', sep='\t')
valid_data.to_csv('Task1a/valid/valid_data.tsv', sep='\t')
```

```python
train_spans = pd.read_csv('Task1b/train/train_spans.tsv', sep='\t', header=None)
train_spans.columns = ['id', 'Class', 'Low', 'High', 'Key']
spans = pd.read_csv('Task1b/valid/spans.tsv', sep='\t', header=None)
spans.columns = ['id', 'Class', 'Low', 'High', 'Key']
```

```python
ids = []
texts = []

with open('Task1b/train/train_tweets.tsv','r') as f:
    for item in f.readlines():
        item = item.strip('\n')
        item_list = item.split('\t')
        assert len(item_list)==2
        ids.append(item_list[0])
        texts.append(item_list[1])
        
train_tweet = pd.DataFrame({'id':ids[0:],
                        'Tweet':texts[0:]
                       })

ids = []
texts = []

with open('Task1b/valid/tweets.tsv','r') as f:
    for item in f.readlines():
        item = item.strip('\n')
        item_list = item.split('\t')
        assert len(item_list)==2
        ids.append(item_list[0])
        texts.append(item_list[1])
        
tweets = pd.DataFrame({'id':ids[0:],
                        'Tweet':texts[0:]
                       })

train_data = pd.merge(train_tweet, train_spans, how='left')
valid_data = pd.merge(tweets, spans, how='left')
train_data.to_csv('Task1b/train/train_data.tsv', sep='\t')
valid_data.to_csv('Task1b/valid/valid_data.tsv', sep='\t')
```

```python
train_spans_norm = pd.read_csv('Task1c/train/train_spans_norm.tsv', sep='\t', header=None)
train_spans_norm.columns = ['id', 'Class', 'Low', 'High', 'Key', 'Norm']
spans_norm = pd.read_csv('Task1c/valid/spans_norm.tsv', sep='\t', header=None)
spans_norm.columns = ['id', 'Class', 'Low', 'High', 'Key', 'Norm']
```

```python
size = len(merge_df.id)
for i in range(5):
    merge_df[int(i*0.2*size):int((i+1)*0.2*size)].to_csv('validation_{}.tsv'.format(i), sep='\t')
    merge_df.drop(list(range(int(i*0.2*size),int((i+1)*0.2*size)))).to_csv('train_{}.tsv'.format(i), sep='\t')
```
